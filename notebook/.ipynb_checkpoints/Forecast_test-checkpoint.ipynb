```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import numpy as np


import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

from statsmodels.tsa.stattools import adfuller


```

```python
ARQ_VENDAS = r'D:\#Mega\Jeferson - Dev\02 - Linguagens\Python\Analise_Producao_Geral\files\dados_vendas.csv'
```

```python
df = pd.read_csv(ARQ_VENDAS)
```

```python
df.head(5)
```

```python
df['Data'] = pd.to_datetime(df['Data'], errors='coerce')
```

```python
df.info()
```

## Preparando o arquivo

```python
#df['Ano'] = df['Data'].dt.year
#df['Mes'] = df['Data'].dt.month
#df['Dia'] = df['Data'].dt.day
```

```python
df['Periodo'] = df['Data'].dt.to_period('M')
```

```python
df.head(5)
```

```python
df_vendas_mes_ano = df.groupby('Periodo')['Quantidade_vendida'].sum().reset_index()
#df_vendas_mes_ano = df.groupby(['Mes', 'Ano'])['Quantidade_vendida'].sum().reset_index()
#df_vendas_mes_ano = df_vendas_mes_ano.sort_values(['Ano', 'Mes'])
```

```python
df_vendas_mes_ano.head(15)
```

```python
df.info()
```

```python
qtd_mensal = df.set_index('Data')['Quantidade_vendida'].resample('ME').sum()
```

```python
qtd_mensal.head(15)
```

## Visualização Inicial da Série Temporal

```python
plt.figure(figsize=(15,6))
qtd_mensal.plot(title='Qtd Vendas Mensal', marker='o', linestyle='-')
plt.xlabel('Data')
plt.ylabel('Quantidade Vendida')
plt.grid(True)
plt.show()
```

## Decomposição da série Temporal

```python
decomposicao = seasonal_decompose(qtd_mensal, model='additive', period=12)
```

```python
fig = decomposicao.plot()
fig.set_size_inches(12, 8)
plt.show()
```
